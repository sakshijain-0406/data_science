```python
#libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
#importing the dataset
wine= pd.read_csv("wine.csv")
wine.head()
```

```python
wine.describe()
```

```python
wine.info()
```

```python
#Slicing the data.we don't require the 1st column 'Type'
Wine=wine.iloc[:,1:]
Wine
```

```python
# converting into array
win_ary=Wine.values
win_ary
```

```python
# normalizing the data
wine_norm= scale(win_ary)
wine_norm
```

```python
#applying PCA
pca=PCA()
pca_values=pca.fit_transform(wine_norm)
pca_values
```

```python
pca=PCA(n_components=6) # here we are taking the no of componenets for PCA to be 6.
pca_values=pca.fit_transform(wine_norm)
```

```python
#PCA components or covariance matrix
pca.components_
```

```python
#the amount of variance that each PCA explains
var=pca.explained_variance_ratio_
var
```

```python
#cumilate the variance
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
#variance plot for PCA components obtained
plt.plot(var1,color="red")
```

```python
#final data frame
final_df= pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
#visualization of PCAs
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df);
```

```python
#plot between PCA1 AND PCA2
x=pca_values[:,0:1] #1st component
y=pca_values[:,1:2]# 2nd components
plt.scatter(x,y)
```

```python
sns.scatterplot(data=final_df,x='PC1',y='PC2',hue='Type')
```

# Hierarchical Clustering (Origonal Data)

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# as we have already normalize the data create dendrogram
plt.figure(figsize=(10,8))
plt.axhline(y=8)
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
#create clusters . taking no.of clusters =3 (from dendrogram)
hcclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hcclusters
```

```python
clusters=pd.DataFrame(hcclusters.fit_predict(wine_norm),columns=['clusters'])
```

```python
#adding clusters to data set
wine2=wine.copy()
wine2['clusterid']=clusters
wine2
```

# K-Means Clustering(Original Data)

```python
from sklearn.cluster import KMeans
```

```python
# As we already have normalized data
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change.
```

```python
# elbow chart
wcss = [] # we have created a list here to store the values(place holder) in a for loop.
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_) # "INERTIA" means within the average distances. "APPEND" means to add
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#build cluaster algorithim and select 3 clusters from the above elbow dig
clusters_new=KMeans(3,random_state=0)
clusters_new.fit(wine_norm)
```

```python
clusters_new.labels_
```

```python
#assign clusters to the dataset
wine3=wine.copy()
wine3['clusters_new']=clusters_new.labels_
wine3
```

#  K-Means Clustering (PCA data, first 3 pca)

```python
# elbow chart
wcss1 = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(final_df[['PC1','PC2','PC3']])
    wcss1.append(kmeans.inertia_) 
    
plt.plot(range(1, 11), wcss1)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
